# Ingestão de dados para RAG com LLM

Este notebook tem como objetivo realizar o processamento e divisão de arquivos Markdown, organizando os textos extraídos para ingestão em uma vector store. O propósito final é habilitar a utilização desses dados em um agente RAG (Retrieval-Augmented Generation), permitindo consultas eficientes e contextualizadas. 

A abordagem inclui:

- **Configuração do modelo LLM**: Utilizando o modelo `deepseek-r1:8b` para interações e processamento de linguagem natural.
- **Ingestão de dados**: Extração e organização dos textos Markdown para armazenamento estruturado.
- **Integração com vector store**: Preparação dos dados para consultas otimizadas em um agente RAG.

Este fluxo garante que os textos sejam processados de forma eficiente, permitindo a criação de agentes inteligentes baseados em recuperação de informações.

### 6.1 . Utilizando LangGraph para monitorar o prompt de entrada, o contexto trazido e a resposta produzida

```python
from langchain_core.documents import Document
from typing_extensions import List, TypedDict, Annotated

from langgraph.graph import START, StateGraph
from pydantic import BaseModel, Field
from langgraph.graph.message import add_messages
from langchain_core.prompts import PromptTemplate


class Query(BaseModel):
    questions: List[str] = Field(
        description="Uma lista de sub questões relacionadas à questão principal."
    )


class State(TypedDict):
    question: str
    query: Query
    context: List[Document]
    answer: str


def analyze_query(state: State):

    # Define the system prompt
    system = """
    Você é um assistente útil que gera sub-perguntas relacionadas a uma pergunta de entrada se for necessário.
    Seu primeiro objetivo é avaliar a necessidade de dividir a pergunta de entrada em um conjunto de no máximo 4 sub-problemas / sub-perguntas
    que possam ser respondidos de forma independente, caso a pergunta seja suficientemente complexa.
    Se a pergunta de entrada for suficientemente simples, você deve retornar uma lista contendo apenas a mensagem original, sem modificações.
    Se a pergunta de entrada for complexa, você deve gerar uma lista de no máximo 4 sub-perguntas relacionadas à pergunta de entrada.
    Caso você retorne mais de uma sub-pergunta, você deve garantir que cada sub-pergunta seja independente e não se sobreponha a outra.
    Além disso, você não deve elaborar perguntas demais. Atenha-se ao que for mais relevante.
    Todas as perguntas devem ser elaboradas sob ótica da fé católica.
    Todas as perguntas devem ser elaboradas imperativamente em português brasileiro (pt-br).
    """

    structured_llm = llm.with_structured_output(Query)
    messages = [SystemMessage(system)] + [HumanMessage(content=state["question"])]
    query = structured_llm.invoke(messages)
    return {"query": query}


def retrieve(state: State):
    questions = state["query"].questions

    retrieved_docs = []
    print(f"Questions: {questions}")
    
    for q in questions:
        retrieved_docs += vector_store.similarity_search(q, k=3)
    return {"context": retrieved_docs}


def generate(state: State):
    docs_content = "\n\n".join(doc.page_content for doc in state["context"])
    prompt = prompt_template.invoke(
        {"question": state["question"], "context": docs_content}
    )
    response = llm.invoke(prompt)
    return {"answer": response.content}
```

```python
graph_builder = StateGraph(State).add_sequence([analyze_query, retrieve, generate])
graph_builder.add_edge(START, "analyze_query")
graph = graph_builder.compile()
```
